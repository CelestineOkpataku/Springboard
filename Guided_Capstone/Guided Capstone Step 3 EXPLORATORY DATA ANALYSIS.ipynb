```python
# to Select categorical Variables only
Data_cat = Data3.select_dtypes(include = 'object').copy()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#count plot for the colume state 
sns.countplot(data = Data_cat, x = 'state')
plt.xticks(rotation =90)
```

```python
# making a barplots for state counts
```

```python
plt.subplots(figsize= (10,10))
x= pd.DataFrame(Data3.state.value_counts())
names= list(x.index)
values= list(x.state)
sns.barplot(x=values, y=names, palette= "RdBu_r")
```

```python
# create the correlation matrix heat map
plt.figure(figsize= (14,14))
sns.heatmap(Data3.corr(), cmap= "YlGnBu")
plt.yticks(rotation=0);
```

```python
# annotating the heatmap and adding line width
plt.figure(figsize= (14,14))
sns.heatmap(Data3.corr(), linewidths= .1, cmap= "YlGnBu", annot=True)
plt.yticks(rotation=0);

```

```python
# making pair plots
g = sns.pairplot(Data3)
```
