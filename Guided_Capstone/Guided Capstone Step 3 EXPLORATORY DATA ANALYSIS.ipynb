```python
# I WILL USE A BOX PLOT TO INVESTIGATE ANOMALITIES AND OUTLIERS
# We will use the folllowing steps
#1. Create A boxplot to check for outliers
#2. Apply outlier removal using the Interquartile Range method IQR
#3. Review how many observations were removed
```

```python
#1.
# Creating boxplot for every column in Data
boxplot = NewData3.boxplot(grid = False, fontsize = 10,vert = False, figsize=(12,15))
```

```python
# The boxplot above shows some outliers in the data so therefor i will employ the INTERQUARTILE RANGE IQR method 
# To deal with the outliers 
```

```python
#2.
# Finding the Interquartile range or IQR that is Q3-Q1
#Finding the first Quartile Q1
Q1 = NewData3.quantile(0.25)
# Finding the 3rd Quartile Q3
Q3= NewData3.quantile(0.75)
# Calculating the interquartile range Q3
IQR = Q3-Q1
IQR.head()
```

```python
lower_limit = Q1-1.5*IQR
upper_limit = Q3+1.5*IQR
```

```python
lower_limit.head()
```

```python
upper_limit.head()
```

```python
NewData3_int= NewData3.astype(int)
NewData3_int.head()
```

```python
NewData3 = NewData3.drop(columns=['Name','state'], axis=1)
NewData3.head()
```

```python
# Removing the outliers from the Dataframe leaving only data without outliers
Data3_no_outliers = NewData3_int[(NewData3_int>lower_limit)&(NewData3_int<upper_limit)]
Data3_no_outliers
```

```python
Data3_no_outliers = NewData3_int[~((NewData3_int < (Q1 - 1.5 * IQR)) |(NewData3_int> (Q3 + 1.5 * IQR))).any(axis=1)]
Data3_no_outliers.head()
```

```python
# Step1. Calling the variable corr_matrix
corr_matrix_Var = Data3_no_outliers.drop(['AdultWeekday','AdultWeekend','daysOpenLastYear','projectedDaysOpen'], axis=1).corr().abs()
corr_matrix_Var
```

```python
# Step2  Getting the upper triangle of the correlation matrix
import numpy as np
upper = corr_matrix_Var.where(np.triu(np.ones(corr_matrix_Var.shape), k=1).astype(np.bool))
upper
```

```python
# Step3. This interatrates through the columns to find and drop any variable that has correlation> 0.95
to_drop = [column for column in upper.columns if any(upper[column] > 0.95)]
to_drop
```
